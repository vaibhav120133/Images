```python
#8
numerical_columns=['Popularity','Duration (ms)']
sns.pairplot(sp[numerical_columns])
```

```python
#9
top_artists = sp.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(10).index
sp_top_artists = sp[sp['Artist'].isin(top_artists)]

# Step 2: Plot a box plot for track duration across different artists
plt.figure(figsize=(12, 6))
sns.violinplot(x='Artist', y='Duration (ms)', data=sp_top_artists)
plt.xticks(rotation=90)  # Rotate artist names for readability
plt.title("Distribution of Track Duration Across Artists (Box Plot)")
plt.show()
```

```python
#10
plt.figure(figsize=(12, 6))
sns.violinplot(x='Artist', y='Popularity', data=sp_top_artists)
plt.xticks(rotation=90)  # Rotate artist names for readability
plt.title("Distribution of Track Popularity Across Artists (Violin Plot)")
plt.show()
```
